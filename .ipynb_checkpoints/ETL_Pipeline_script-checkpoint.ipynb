```python
spark = SparkSession.builder.appName("Advanced_ETL_Pipeline").getOrCreate()
```

```python
# ============================================
# 1. BRONZE LAYER (RAW INGESTION)
# ============================================

orders_df = spark.read.option("header", True).csv("orders.csv")
customers_df = spark.read.option("header", True).csv("customers.csv")
products_df = spark.read.option("header", True).csv("products.csv")
```

```python
# Add ingestion timestamp
orders_df = orders_df.withColumn("ingestion_time", current_timestamp())
customers_df = customers_df.withColumn("ingestion_time", current_timestamp())
products_df = products_df.withColumn("ingestion_time", current_timestamp())
```
